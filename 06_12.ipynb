```python
# @classmethod, @staticmethod의 기능과 차이: python에서 정적함수 
# 장식자차이 
# @classmethod: self대신 cls(자체 클래스)가 self역할
# @staticmethod: 매개변수에 self없음
# 인스턴스하지 않아도 호출가능
import datetime

class Person(object):
    my_class_var = '에이콘'
    def __init__(self, year, month, day, sex): #상속되지 않음, 오버로딩 불가
        self.year = year
        self.month = month
        self.day = day
        self.sex = sex
        
    def __str__(self):
        return '{}년 {}월 {}일 생 {}입니다.'.format(self.year, self.month, self.day, self.sex)
    
    @classmethod # 초기화 다양성 가능, 클래스 자체를 받음
    def ssn_constructor(cls, ssn):
        front, back = ssn.split('-') #기호를 중심으로 문자열 분리
        sex = back[0]
        if sex == '1' or sex == '2':
            year = '19' + front[:2]
        else:
            year = '20' + front[:2]
        if(int(sex) % 2) == 0:
            sex='여성'
        else:
            sex='남성'
            
        month = front[2:4]
        day = front[4:6]
        return cls(year, month, day, sex) # 클래스 초기화
    
    @staticmethod
    def is_work_day(day):
        if day.weekday() == 5 or day.weekday() == 6:
            return False
        return True
    
ssn_1 = '900829-1034356'
ssn_2 = '051224-4061569'
person_1 = Person.ssn_constructor(ssn_1)
print(person_1)
person_2 = Person.ssn_constructor(ssn_2)
print(person_2)
my_date = datetime.date(2016, 10, 9)
print(Person.is_work_day(my_date))
print(person_1.is_work_day(my_date))
```

```python
#상속
class Parent:
    parentAttr = 100
    def __init__(self):
        print("부모 생성자")
    def parentMethod(self):
        print("부모 메소드")
    def setAttr(self, attr):
        Parent.parentAttr = attr
    def getAttr(self):
        print("부모 속성 :", Parent.parentAttr)
        
class Child(Parent):
    def __init__(self): # 부모의 생성자는 명시적으로 호출 가능하다
        Parent.__init__(self) # 부모의 생성자는 상속되지도 재정의도 안된다.
        print("자식 생성자")
    def childMethod(self):
        print("자식 메소드")
    def print_attribute(self):
        self.setAttr(500)
        print(" 자식 호출된 부모 속성 :", Parent.parentAttr)
        
c = Child()
c.childMethod()
c.parentMethod()
c.setAttr(200)
c.getAttr()
c.print_attribute()
```

```python
# delegation(위임) : 상속구조와 대표타입 프로그래밍을 이용해서 변화에 대응하는 문제해결방식

class Animal:
    def Name(self):
        pass
    def Sleep(self):
        print('sleep')
    def MakeNoise(self):
        pass
    
class Dog(Animal):
    def Name(self):
        print("I'm dog")
    def MakeNoise(self):
        print("woof!")

# 고양이, 사자
class TestAnimals:
    def PrintName(self, animal): # 매개변수로 받아들이는 클래스에 작업 위임
        animal.Name()
    def MakeNoise(self, animal):
        animal.MakeNoise()
        
TestAnimals = TestAnimals()
dog = Dog()
TestAnimals.PrintName(dog)
TestAnimals.MakeNoise(dog)
```

```python
class Grade:
    def scoring(self, grade):
         raise NotImplementedError
        
class ABC(Grade):
    def __init__(self, sc):
        self.score = sc
        
    def scoring(self):
        if(self.score > 90):
            return 'A학점'
        elif(self.score > 80):
            return 'B학점'
        else:
            return 'C학점'

class PF(Grade):
    def __init__(self, sc):
        self.score = sc
        
    def scoring(self):
        if(self.score > 80):
            return "Pass"
        else:
            return "Fail"
class SW(Grade):
    def __init__(self, sc):
        self.score = sc
    def scoring(self):   
        if(self.score > 90):
            return '수'
        elif(self.score > 80):
            return '우'
        else:
            return '미'
        
class printer():
    def printresult(self, t):
        print(t.scoring())
        
abc = ABC(100)
abc2 = ABC(90)
pf = PF(100)
pf2 = PF(70)
sw = SW(100)
sw2 = SW(90)
p = printer()
p.printresult(abc)
p.printresult(abc2)
p.printresult(pf)
p.printresult(pf2)
p.printresult(sw)
p.printresult(sw2)
```

```python
# 삼각형, 사각형, 원의 면적과 둘레를 관리하는 프로그램을 작성
# 10개의 객체를 관리하고자 하는데 이를 편리한 구조로 관리할 수 있도록 프로그램하고 객체를 리스트로 관리하여 반복문 하나로 관리하는
# 프로그램 작성하시오
import random

class Shape:
    def area(self):
        raise NotImplementedError

class Tri(Shape):
    def __init__(self, a, b):
        self.under = a
        self.height = b
        
    def area(self):
        print("삼각형 넓이: " + str(self.under * self.height / 2))
    
class Rect(Shape):
    def __init__(self, a, b):
        self.width = a
        self.height = b
        
    def area(self):
        print("사각형 넓이: " + str(self.width * self.height))
    
shape_list = []
for i in range(1, 6):
    rand = random.randint(1,100)
    rand2 = random.randint(1,100)
    shape_list.append(Tri(rand, rand2))
    shape_list.append(Rect(rand, rand2))
    
for i in range(10):
    shape_list[i].area()
    #shape_list[i].perimeter()
```

```python
# strategy pattern
def lower(str):
    print(str.lower())

def upper(str):
    print(str.upper())
    
class Person(object): # 프로그램을 변경하지 않고 행동을 변화하는 것이 가능
    def __init__(self, func = None): # 행동함수의 결정이 초기화에 의해 결정
        if func is not None:
            self.exe = func
    def exe(self, name):
        print(name)
        
lw_person = Person(lower)
lw_person.exe('Kim jongho')
up_person = Person(upper)
up_person.exe('Korea')
```

```python
# singleton pattern
# __call__:  
# __new__: 생성자보다 먼저 실행
# __init__: 생성자
# __dict__: 매개변수와 값

class Person(object):
    __instance = None
    def __new__(cls, *args, **kwargs):
        print("실행")
        if not cls.__instance:
            cls.__instance = super(Person, cls).__new__(cls)
        return cls.__instance
    
    def __init__(self, name):
        self.name = name

tom = Person("초등학교")
Jane = Person("대학교")

print(tom is Jane)
print(tom.name)
print(Jane.name) # 주소는 동일하고 값은 변경
```

```python
# adpater pattern
class Target(object):
    def specific_request(self):
        return 'Hello Adapter Pattern!'

class Adapter(object):
    def __init__(self, adaptee):
        self.adaptee = adaptee

    def request(self):
        return self.adaptee.specific_request()

client = Adapter(Target())
print(client.request())
```

```python
class Clock(object):
    def __init__(self, hours=0, minutes=0,seconds=0):
        self.__hours = hours
        self.__minutes = minutes
        self.__seconds = seconds
        
    def set(self, hours, minutes, seconds=0):
        self.__hours = hours
        self.__minutes = minutes
        self.__seconds = seconds
        
    def tik(self):
        flag = 0
        if self.__seconds == 59:
            self.__seconds = 0
            if self.__minutes == 59:
                self.__minutes = 0
                if self.__hours == 23:
                    self.__hours = 0
                    flag = 1
                else:
                    self.__hours += 1
            else:
                self.__minutes += 1
        else:
            self.__seconds += 1
        return flag
    
    def display(self):
        print("%d:%d:%d" % (self.__hours, self.__minutes, self.__seconds))
        
    def __str__(self):
        return "%2d:%2d:%2d" % (self.__hours, self.__minutes, self.__seconds)
    
x = Clock()
print(x)
for i in range(43200):
    x.tik()
print(x)                   
```

```python
class Calendar(object):
    month = (31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31)
    
    def __init__(self, day = 1, month = 1,year = 2100):
        self.__day = day
        self.__month = month
        self.__year = year
        
    def leapyear(self, y):
        if y % 4:
            return 0
        else:
            if y % 100:
                return 1
            else:
                if y % 400:
                    return 0
                else:
                    return 1
                
    def set(self, day, month, year):
        self.__day = day
        self.__month = month
        self.__year = year
        
    def get():
        return (self, self.__day, self.__month, self.__year)
    
    def advance(self):
        months = Calendar.month
        max_days = months[self.__month - 1]
        
        if self.__month == 2:
            max_days += self.leapyear(self.__year)
            
        if self.__day == max_days:
            self.__day = 1
            if self.__month == 12:
                self.__month = 1
                self.__year += 1
            else:
                self.__month += 1
        else:
            self.__day += 1

    def __str__(self):
        return str(self.__day) + "/" + str(self.__month) + "/" + str(self.__year)
    
x = Calendar()
print(x)
for i in range(58):
    x.advance()
print(x)
```

```python
class CalendarClock(Clock, Calendar):
    def __init__(self, year = 1900, month = 1, day = 1, hours = 0, minutes = 0, seconds = 0):
        Calendar.__init__(self, day, month, year)
        Clock.__init__(self, hours, minutes, seconds)
    
    def __str__(self):
        return Calendar.__str__(self) + ", " + Clock.__str__(self)
    
    def tick(self):
        if self.tik() == 1:
            self.advance()
        
x = CalendarClock()
print(x)
for i in range(86400):
    x.tick()
print(x)
```

```python
from datetime import datetime

class Calendar(object):
    months = (31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31)
    
    def __init__(self, year = 2000, month = 1,day = 1):
        self.__day = day
        self.__month = month
        self.__year = year
        
    def __sub__(self, other):
        gap = 0
        
        other_year, other_month, other_day = other.get()
        max_day = self.months[other_month-1]
        
        gap += (self.__year - other_year) * 365
        for i in range(max_day, max_day + self.__month - other_month):
            gap += self.months[i%12]
        print()
        print(gap)
        print()
        gap += self.__day - other_day

        return gap
                
    def set(self, day, month, year):
        self.__day = day
        self.__month = month
        self.__year = year
        
    def get(self):
        return (self.__year, self.__month, self.__day)

    def __str__(self):
        return str(self.__day) + "/" + str(self.__month) + "/" + str(self.__year)
    
#######################################################################################################################

class Item:
    def __init__(self, name, buy, exp):
        self.name = name
        self.buy = buy
        self.exp = exp
        self.rem = self.expireChk()
        
    def __repr__(self):
        if self.rem < 0:
            return "품목이름: %s\t구입일: %s\t유통기한: %s\t잔여기일: 유통기한지남" % (self.name, self.buy, self.exp)
        else:
            return "품목이름: %s\t구입일: %s\t유통기한: %s\t잔여기일: %s" % (self.name, self.buy, self.exp, self.rem)
    
    def expireChk(self):
        today = Calendar(datetime.today().year, datetime.today().month, datetime.today().day)
        print("*******************"+ str(today - self.exp))
        return today - self.exp
    
#######################################################################################################################

class Management:
    items = []
    
    def add(self, name, buy, exp):
        self.items.append(Item(name, buy, exp))
     
    def printall(self):
        for i in range(len(self.items)):
            print(self.items[i])
            
    def delExp(self):
        for i in range(len(self.items)):
            if self.items[i].rem < 0:
                del self.items[i]

#######################################################################################################################

def main():
    management = Management()
    
    while True:
        select = input("1.품목 입력 2.출력 3.품목제거 4.종료   ")
        if select == '1':
            name = input("품목이름: ")
            buy = input("구입일: ")
            year, month, day = buy.split(' ')
            buy_cal = Calendar(int(year), int(month), int(day))
            exp = input("유통기한: ")
            year, month, day = exp.split(' ')
            exp_cal = Calendar(int(year), int(month), int(day))
            management.add(name, buy_cal, exp_cal)
            
        elif select == '2':
            management.printall()
            
        elif select == '3':
            management.delExp()
            
        else:
            print("종료")
            break
            
main()
```
