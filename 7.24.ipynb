```python
import matplotlib as mpl

import matplotlib.pyplot as plt
import numpy as np
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
# Alias -> Property
# 'lw' -> 'linewidth'
# 'ls' -> 'linestyle'
# 'c' -> 'color'
# 'fc' -> 'facecolor'
# 'ec' -> 'edgecolor'
# 'mew' -> 'markeredgewidth'
# 'aa' -> 'antialiased'
# mpl.rc('lines', linewidth = 2, color='r')
mpl.rc('font', family = 'NanumGothic')
# font = {'family' : 'monospace',
#          'weight' : 'bold',
#          'size' : 'larger')
#  mpl.rc('font', **font) # dictionary 로 여러개의 매개변수 전달시 사용
mpl.rc('axes', unicode_minus=False)
plt.title('한글 제목')
plt.plot(X, C, label = '코사인')
t = 2 * np.pi/3
plt.scatter(t, np.cos(t), 50, color='blue') #  한 점을 출력
plt.xlabel("엑스축 라벨")
plt.ylabel("와이축 라벨")
plt.annotate("여기가 0.5!", xy=(t, np.cos(t)), xycoords='data',
            xytext=(-90, -50),
            textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle="->"))
plt.legend(loc=2)
plt.show()
```

```python
# 3차원 : wireframe - 선, surface 면 , 체
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
# row step, column step 간격
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap = 'hot')

plt.show()
```

```python
# 데이터에 함수적용
# 데이터프레임 단위로 적용
# apply(행열방향으로), applymap(원소별로), map(하나의 열을 기준으로)의 차이

frame = pd.DataFrame((np.random.randn(4,3)), columns = ['rain', 
        'income', 'tax'], index = ['seoul','daejun','incheon','daegu'])
print(frame)
print("absolute 함수 적용", np.abs(frame))
f=lambda x: x.max() - x.min() # range
print("함수 객체의 행 적용 (열방향)", frame.apply(f))
print("함수 객체 열 적용 (행방향)", frame.apply(f, axis = 1))

def f(x):
    return pd.Series([x.min(), x.max()], index=['min', 'max'])

print("함수객체 적용", frame.apply(f))
format=lambda x: '%.2f' %x
print(frame.applymap(format))
print(frame['rain'].map(format))
```

```python
# index_col : 특별한 컬럼의 내용을 인덱스로 사요하고자 할 때 사용
# chunksize : 대용량 데이터를 메모리에 로딩할 때 일정한 사이즈로 나누어서
# 로딩하고자 할 때 사용 로딩이후 concat 를 사용하여 데이터를 결합
import pandas as pd
import matplotlib.pyplot as plt
import os
bigFilePath = "weather_2012.csv"

chunker = pd.read_csv(bigFilePath, chunksize=1000, index_col='Date/Time',
                     encoding="UTF-8")
weather_2012_final = pd.concat([x for x in chunker], ignore_index = True)
print(weather_2012_final.shape)
weather_2012_final.head(3)

print(weather_2012_final)

weather_2012_final = weather_2012_final.dropna(axis=1, how = 'any')
print(weather_2012_final[:5])
print(weather_2012_final.describe())
weather_2012_final['Temp (C)'].plot(figsize=(30, 12))
weather_2012_final.boxplot()
print(os.getcwd())
```
