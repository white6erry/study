```python
# Point class: x,y 두 포인터간의 거리값을 구하시오
import math

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def dist(self, other):
        return math.sqrt((self.x - other.x)**2 + (self.y - other.y)**2)
    def __repr__(self):
        return "좌표는 (%d, %d)" % (self.x, self.y)
    
point1 = Point(7,3)
point2 = Point(4,7)
print(point1)
print(point2)
print("두 점사이 거리:", point1.dist(point2))
```

```python
# 문제 Point클래스를 이용해서 원 class를 정의하시오
class Circle:
    def __init__(self, p, r):
        self.p = p
        self.r = r
    def dist(self, other):
        return self.p.dist(other.p)
    def isCrash(self, other):
        if self.dist(other) <= self.r + other.r:
            return "충돌함"
        else:
            return "충돌 안함"
    def __repr__(self):
        return str(self.p) + " 반지름은 %d" % self.r
    
cir1 = Circle(point1, 2)
cir2 = Circle(point2, 5)
print(cir1)
print(cir2)
print(cir1.isCrash(cir2))
```

```python
# 문제 Rectangle을 두 point클래스를 이용하여 정의하시오
# 사각형의 면적, 둘레, 대각선의 길이를 구하는 함수작성
class Rectangle:
    def __init__(self, p1, p2):
        self.p1 = p1
        self.p2 = p2
    def area(self):
        return abs(self.p1.x - self.p2.x) * abs(self.p1.y - self.p2.y)
    def perimeter(self):
        return 2 * (abs(self.p1.x - self.p2.x) + abs(self.p1.y - self.p2.y))
    def diagonal_dist(self):
        return self.p1.dist(self.p2)
    
r = Rectangle(point1, point2)
print(r.area())
print(r.perimeter())
print(r.diagonal_dist())
```

```python
# 실시간 매개변수, 함수 추가

class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second
    def sum(self):
        result = self.first + self.second
        return result
    def mul(self):
        result = self.first * self.second
        return result
    def sub(self):
        result = self.first - self.second
        return result
    def div(self):
        result = self.first / self.second
        return result
    def all_call(self):
        result1 = self.sum()
        result2 = self.mul()
        result3 = self.sub()
        result4 = self.div()
        return result1+result2+result3+result4

cal = FourCal()
cal.setdata(100, 200)
# 실시간 변수 추가
cal.first2 = 200
cal.second2 = 300

def addnum(a, b):
    return a + b
# 실시간 함수추가(인스턴스화한 모든 object에 영향을 미침)
cal.addnum_point = addnum
print(cal.addnum_point(cal.first, cal.second2))
print(cal.all_call())
```

```python
# 3차원 벡터 클래스 생성하기
class Vector:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
    def __repr__(self):
        return "(좌표는 %d, %d, %d) " % (self.x, self.y, self.z)
    def __add__(self, other): # 연산자 오버로딩
        return Vector(self.x + other.x, self.y + other.y, self.z + other.z)
    def __mul__(self, other):
        return Vector(self.x * other.x, self.y * other.y, self.z * other.z)
    def norm(self): # 벡터의 크기
        return math.sqrt(self.x**2 + self.y**2 + self.z**2)
    def dot(self, other): # 내적
        return (self.x * other.x + self.y * other.y + self.z * other.z) / (self.norm() * other.norm())

f = Vector(10, 0, 0)
g = Vector(0, 27, 0)
# 직교할 때 내적 = 0
print(f.dot(g))
print(f + f)
print(f + g)
print(f * g)
print(f.norm())
```

```python
#클래스를 매개변수로 받는 함수

class Guard:
    def three_point(self):
        print("가드는 3점슛 장착")
    def apass(self):
        print("가드의 핵심은 에이패스지")
        
class Jordan:
    def three_point(self):
        print("조단은 3점슛 장착")
    def apass(self):
        print("조단은 에이패스 가능")
        
def inhecourt(duck): # 매개변수 클래스
    duck.three_point()
    duck.apass()
    
def game():
    player_a = Guard()
    player_b = Jordan()
    inhecourt(player_a)
    inhecourt(player_b)
game()    
```

```python
# class로 만든 장식자
class Verbose:
    def __init__(self,f):
        print("초기화")
        self.func = f
    def __call__(self):
        print("데코레이터 시작", self.func.__name__)
        self.func()
        print("데코레이터 종료", self.func.__name__)
        
# @Verbose를 마주치면 생성자를 실행한다. 때문에 초기화가 가장 먼저 출력됨
@Verbose
def my_function():
    print("원래함수의 기능")
    
print("프로그램 시작")
my_function()
```

```python
# 성적 프로그램의 메뉴화
# Student 클래스
# 초기화, 입력함수, 합계와 평균함수, level함수, __repr__(1사람 출력), __cmp__
# Management클래스
# 추가함수, 수정함수, 검색함수, 삭제함수, 전체 출력
# 메뉴화
# 번호를 이용하여 검색하고 이름을 변경하는 메뉴를 추가하시오
# 국어 영어 수학 점수를 선택해서 변경하는 메뉴를 추가
# 유학지원 가능 여부를 확인하는 메뉴를 추가하여 총점 90점이상인 학생은 유학가능 미만은 유학불가로 출력
```

```python
class Student:
    def __init__(self, name, kor, eng, mat, number):
        self.name = name
        self.num = number
        self.kor = kor
        self.eng = eng
        self.mat = mat
        
    def cal(self):
        self.tot = self.kor + self.eng + self.mat
        self.avg = self.tot / 3    
        if self.avg >= 90:
            self.support = 1
        else:
            self.support = 0
        
    def level(self):
        if self.tot > 290:
            self.lv = 'A'
        elif self.tot > 270:
            self.lv = 'B'
        elif self.tot > 250:
            self.lv = 'C'
        elif self.tot > 230:
            self.lv = 'D'
        else:
            self.lv = 'F'
        return self.lv
    
    def __repr__(self):
        return  "번호: %d 이름:%s 국어:%s 영어:%s 수학:%s 총점:%s 평균:%.2f 학점:%s" % (self.num, self.name, self.kor, self.eng, self.mat, self.tot, self.avg, self.lv)
    
    def __cmp__(self, other):
        if self.tot < other.tot:
            return 1
        elif self.tot == other.tot:
            return 0
        else:
            return -1
        
    def inputData(self, kor, eng, mat):
        self.kor = kor
        self.eng = eng
        self.mat = mat
        
        
class Management:
    students = []
    student_num = 0
    
    def addStu(self, s):
        self.student_num += 1
        self.students.append(s)
        self.students.sort(key = lambda x: x.tot, reverse = True)
                
    def modKor(self, name, kor):
        num = self.search(name)
        self.students[num].kor = kor
        print("변경된 점수")
        print(self.students[num])
        
    def modEng(self, name, eng):
        num = self.search(name)
        self.students[num].eng = eng
        print("변경된 점수")
        print(self.students[num])
        
    def modMat(self, name, mat):
        num = self.search(name)
        self.students[num].mat = mat
        print("변경된 점수")
        print(self.students[num])
        
    def modScore(self, name, kor, eng, mat):
        num = self.search(name)      
        self.students[num].kor = kor
        self.students[num].eng = eng
        self.students[num].mat = mat
        print("변경된 점수")
        print(self.students[num])
        self.students[num].cal
        
    def search(self, name):
        for i in range(0, self.student_num):
            if name == self.students[i].name:
                print(self.students[i])
                return i
        print("찾는 학생이 없음")    
        return -1
    
    def searchNum(self, num):
        for i in range(0, self.student_num):
            if num == self.students[i].num:
                print(self.students[i])
                return i
            
    def delStudent(self, name):
        num = self.search(name)
        if num == -1:
            print("해당 학생 없음")
        else:
            del self.students[num]
            print("삭제완료")
        self.student_num -= 1
        
    def print(self):
        print("총 학생수 %d" % self.student_num)
        for i in range(self.student_num):
            print(self.students[i])
    
    def isSupport(self, name):
        if self.students[self.search(name)].support == 1:
            print("유학가능")
        else:
            print("유학불가")

def start(manager):
    while True:
        print()
        choice = input("1.신규 입력 2.점수 수정 3.검색 4.삭제 5.전체출력 6. 번호로 검색 7.점수별 수정 8.유학지원 확인 9.종료 =>")
        print()
        if choice == '1':
            name = input("이름: ")
            kor = int(input("국어점수: "))
            eng = int(input("영어점수: "))
            mat = int(input("수학점수: "))
            num = int(input("번호: "))
            student = Student(name, kor, eng, mat, num)
            student.cal();student.level()
            manager.addStu(student)
            
        elif choice == '2':
            name = input("수정할 이름: ")
            kor = int(input("국어점수: "))
            eng = int(input("영어점수: "))
            mat = int(input("수학점수: "))
            manager.modScore(name, kor, eng, mat);
            
        elif choice == '3':
            manager.search(input("검색할 이름: "))
            
        elif choice == '4':
            manager.delStudent(input("삭제할 이름: "));
            
        elif choice == '5':     
            manager.print()
        
        elif choice == '6':     
            manager.searchNum(int(input("검색할 번호: ")))
            
        elif choice == '7':    
            name = input("수정할 이름: ")
            select = input("국어, 영어 수학중 선택:")
            if select == '국어':
                manager.modKor(name, int(input("점수입력: ")))
            if select == '영어':
                manager.modEng(name, int(input("점수입력: ")))
            if select == '수학':
                manager.modMat(name, int(input("점수입력: ")))
                
        elif choice == '8':     
            manager.isSupport(input("확인할 이름: "))
            
        elif choice == '9':
            break

manager = Management()
start(manager)         
```
